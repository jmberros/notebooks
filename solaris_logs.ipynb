# Apache2 & Nginx logs

```python
import pandas as pd

def read_logs(filenames, fields=[]):
    frames = [pd.read_table(fn, sep='\s+', names=fields) for fn in filenames]
    df = pd.concat(frames)
    if 'timestamp' in df:
        df.timestamp = pd.to_datetime(df.timestamp, format='[%d/%b/%Y:%H:%M:%S')
        df.set_index('timestamp', inplace=True)
        df.drop('tz', axis=1, inplace=True)
    return df
```

```python
filenames = pj.data_files('apache2/access*')
fields = ('host remote_logname remote_user timestamp tz request status '
          'size_of_response referer user_agent').split()

apache_access = read_logs(filenames, fields)

filenames = pj.data_files('nginx/access*')
fields = ('remote_addr - remote_user timestamp tz request status bytes_sent '
          'http_referer http_user_agent').split()

nginx_access = read_logs(filenames, fields)
```
